```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Subject settings
subjects1 = ['toan', 'ngu_van', 'ngoai_ngu']
subject_name = ['Toán', 'Ngữ Văn', 'Ngoại Ngữ']
colors1 = ['skyblue', 'orange', 'olive']

# Define bin widths per subject
bin_widths = {'toan': 0.2, 'ngoai_ngu': 0.2, 'ngu_van': 0.25}

# Set up the figure: 5 rows (years), len(subjects) columns
fig, axes = plt.subplots(nrows=5, ncols=len(subjects1), figsize=(4 * len(subjects1), 20), sharex='col')

# Plotting each distribution
for i, (year, df) in enumerate(data_by_year.items()):
    for j, subject in enumerate(subjects1):
        ax = axes[i, j]

        # Define bin width and create bins
        bin_width = bin_widths[subject]
        bins = [round(x * bin_width, 2) for x in range(int(10 / bin_width) + 1)]

        # Round scores to appropriate interval and count
        rounded_scores = (df[subject].dropna() / bin_width).round().mul(bin_width).round(2)
        score_counts = rounded_scores.value_counts().sort_index()

        # Ensure all bins are represented
        full_index = pd.Series(0, index=bins)
        score_counts = full_index.add(score_counts, fill_value=0)

        # Prepare DataFrame for plotting
        plot_df = pd.DataFrame({
            'score': score_counts.index,
            'count': score_counts.values
        })

        sns.barplot(
            data=plot_df,
            x='score',
            y='count',
            color=colors1[j],
            ax=ax
        )

        # Set title for top row
        if i == 0:
            ax.set_title(subject_name[j], fontsize=10)

        # Year label for left column
        if j == 0:
            ax.annotate(f"{year}", xy=(0, 0.5), xycoords="axes fraction", fontsize=10,
                        ha="center", va="center", rotation=90)

        # Remove x-axis labels and ticks
        ax.set_xlabel('')
        ax.tick_params(axis='x', bottom=False, labelbottom=False)

        # Remove y-axis labels (count)
        ax.set_ylabel('')

        # Show grid
        ax.grid(True, which='both', linestyle='--', linewidth=0.5)

# Layout adjustment
plt.tight_layout()
plt.suptitle("Distribution of Subject Scores (2020–2024)", fontsize=18, y=1.02)
plt.show()
```

```python
from scipy.interpolate import make_interp_spline

subject_means = df20_24.groupby('nam_thi')[subjects].mean().reset_index()

label_subject = ['Toán', 'Ngữ Văn', 'Vật Lý', 'Hóa Học', 'Sinh Học',
                  'Lịch Sử', 'Địa Lý', 'GDCD', 'Ngoại Ngữ']

# Interpolate for smooth curves
plt.figure(figsize=(10, 6))
for subject, label in zip(subjects, label_subject):
    x = subject_means['nam_thi']
    y = subject_means[subject]
    x_new = np.linspace(x.min(), x.max(), 300)  # Generate more x-points
    spline = make_interp_spline(x, y)
    y_smooth = spline(x_new)
    plt.plot(x_new, y_smooth, label=label)

# Customize the plot
plt.title('Mean of Scores by Subjects (2020–2024)')
plt.xlabel('')
plt.ylabel('')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(ticks = subject_means['nam_thi'], labels=subject_means['nam_thi'])
plt.tight_layout()
plt.show()
```

```python

# Define benchmark for A00
benchmarks_A00 = {
    2024: 28.18,
    2023: 27.6,
    2022: 28.25,
    2021: 28.55,
    2020: 28.15
}

# Function to get benchmark for a block and year
def get_benchmark(block, year):
    base = benchmarks_A00.get(year)
    if base is None:
        return None
    return base if block == 'A00' else base - 0.5

# Melt DataFrame to long format
blocks = ['A00', 'A01', 'D01', 'D07']
df_long = df_block.melt(id_vars=['sbd', 'nam_thi'], value_vars=blocks,
                  var_name='block', value_name='mark')

# Drop missing marks
df_long = df_long.dropna(subset=['mark'])

# Add benchmark column
df_long['benchmark'] = df_long.apply(lambda row: get_benchmark(row['block'], row['nam_thi']), axis=1)

# Filter out rows without a benchmark
df_long = df_long[df_long['benchmark'].notna()]

# Determine if mark is above benchmark
df_long['above_benchmark'] = df_long['mark'] > df_long['benchmark']

# Filter only those who passed the benchmark
df_passed = df_long[df_long['above_benchmark']]

# Count qualifying candidates per year/block
pivot = pd.pivot_table(df_passed, index='nam_thi', columns='block',
                       values='sbd', aggfunc='count', fill_value=0)

# Display pivot
pivot
```
